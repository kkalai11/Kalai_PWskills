```python
# # Q1. How can you create a Bokeh plot using Python code?
# we can create a Bokeh plot using Python code by importing the necessary Bokeh modules,
# creating a figure, adding glyphs (like lines or circles), and then displaying the plot.
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Show plots inside Jupyter Notebook (optional)
output_notebook()

# Create some sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a new plot with title and axis labels
p = figure(title="Simple Line Plot", x_axis_label='X-Axis', y_axis_label='Y-Axis')

# Add a line renderer with legend and line thickness
p.line(x, y, legend_label="Line", line_width=2)

# Show the plot
show(p)

```

```python
# Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

# In Bokeh, glyphs are the basic visual building blocks — they are the shapes or marks (like circles, lines, bars, rectangles, etc.) that represent your data on the plot.
# Glyphs = visual representation of data points (like "markers" in Matplotlib or "marks" in Vega).
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Show plot in notebook
output_notebook()

# Sample data
x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

# Create a figure
p = figure(title="Circle Glyph Example", x_axis_label='X', y_axis_label='Y')

# Add circle glyphs
p.circle(x, y, size=10, color="navy", alpha=0.5)

# Show plot
show(p)

```

```python
# Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

from bokeh.plotting import figure, show
from bokeh.models import NumeralTickFormatter, HoverTool
import numpy as np

# Sample data
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Create figure
p = figure(width=800, height=400, tools="pan,wheel_zoom,box_zoom,reset")

# Add data series
line1 = p.line(x, y1, legend_label="sin(x)", line_width=2, color="navy")
line2 = p.line(x, y2, legend_label="cos(x)", line_width=2, color="firebrick", line_dash="dashed")

# Title
p.title.text = "Sine and Cosine Functions"
p.title.text_font_size = '16pt'
p.title.text_font_style = 'bold'
p.title.align = 'center'

# Axes
p.xaxis.axis_label = "X Values"
p.yaxis.axis_label = "Y Values"
p.axis.axis_label_text_font_style = 'italic'
p.axis.axis_label_text_font_size = '12pt'

# Grid
p.grid.grid_line_alpha = 0.3
p.grid.grid_line_dash = [6, 4]

# Legend
p.legend.location = "top_right"
p.legend.background_fill_alpha = 0.7
p.legend.border_line_width = 1
p.legend.click_policy = "hide"

# Plot area
p.background_fill_color = "#f8f9fa"
p.border_fill_color = "white"
p.min_border = 40

# Add hover tool
hover = HoverTool(tooltips=[("value", "@y{0.000}")], mode="vline")
p.add_tools(hover)

show(p)
```
