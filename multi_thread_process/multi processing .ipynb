```python
import multiprocessing
import time  # We need this for the delay

def producer(q):
    for i in ["check", "this", "a", "b", "d"]:
        q.put(i)
        print(f"this is putting {i}: {i}")
    q.put(None)  # Add None to signal the consumer to stop

def consumer(q):
    while True:
        item = q.get()
        if item is None:
            break
        print(f"this is getting {item}: {item}")

if __name__ == "__main__":
    Q = multiprocessing.Queue()
    
    # Start producer process
    m1 = multiprocessing.Process(target=producer, args=(Q,))
    m1.start()
    
    # Add a small delay to ensure the producer has time to put items in the queue
    time.sleep(1)  # Adjust the sleep time if necessary
    
    # Start consumer process
    m2 = multiprocessing.Process(target=consumer, args=(Q,))
    m2.start()

    # Wait for both processes to finish
    m1.join()
    m2.join()
```

```python
import multiprocessing

def producer(q, event):
    for i in ["check", "this", "a", "b", "d"]:
        q.put(i)
        print(f"this is putting {i}: {i}")
    q.put(None)  # Add None to signal the consumer to stop
    event.set()  # Signal that the producer is done

def consumer(q, event):
    event.wait()  # Wait for the signal from producer
    while True:
        item = q.get()
        if item is None:
            break
        print(f"this is getting {item}: {item}")

if __name__ == "__main__":
    Q = multiprocessing.Queue()
    event = multiprocessing.Event()  # Event to synchronize processes
    
    m1 = multiprocessing.Process(target=producer, args=(Q, event))
    m2 = multiprocessing.Process(target=consumer, args=(Q, event))
    
    m1.start()  # Start the producer
    m2.start()  # Start the consumer
    
    m1.join()  # Wait for producer to finish
    m2.join()  # Wait for consumer to finish
```

```python
#piping 

import multiprocessing

def sender(conn,msg):
    for i in msg:
        conn.send(i)
    conn.close()
    
def receive(conn):
    while True:
        try:
            msg=conn.recv()
        except Exception as e:
            print(e)
            break
        print(msg)

if __name__ == "__main__":
 msg= ["hello","multiprocessing","world","python"]
 parent_conn, child_conn = multiprocessing.Pipe()
 m1=multiprocessing.Process(target=sender,args=(parent_conn,msg)) 
 m2 = multiprocessing.Process(target=receive,args=(child_conn,))  
 m1.start()
 m2.start()
 m1.join()
 parent_conn.close()
 m2.join()
 child_conn.close()


```
