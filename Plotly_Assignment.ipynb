```python
# Q1. Load the "titanic" dataset using the load_dataset function of seaborn.
# Use Plotly express to plot a # scatter plot for age and fare columns in the titanic dataset.
```

```python
import plotly.express as px
import seaborn as sns
import pandas as pd
import plotly.graph_objects as go

# Load Titanic dataset from seaborn
titanic = sns.load_dataset('titanic')

# Drop rows with missing values in 'age' or 'fare'
titanic_cleaned = titanic.dropna(subset=['age', 'fare'])

# Create scatter plot
fig = px.scatter(
    titanic_cleaned,
    x='age',
    y='fare',
    title='Titanic: Age vs Fare',
    labels={'age': 'Age', 'fare': 'Fare'},
    hover_data=['sex', 'class', 'embarked']
)

# Show plot
fig.show()



```

```python
# Q2. Using the tips dataset in the Plotly library, plot a box plot using Plotly express.
# Load tips dataset from seaborn
tips = sns.load_dataset('tips')

# Create box plot
fig = px.box(
    tips,
    x='day',
    y='total_bill',
    color='sex',  # Optional: adds a color split by gender
    title='Box Plot of Total Bill by Day',
    labels={'total_bill': 'Total Bill', 'day': 'Day of the Week'}
)

# Show the plot
fig.show()
```

```python
# prompt: Q3. Using the tips dataset in the Plotly library, Plot a histogram for x= "sex" and y="total_bill" column in
# the tips dataset. Also, use the "smoker" column with the pattern_shape parameter and the "day"
# column with the color parameter.

fig = px.histogram(tips,
                   x="sex",
                   y="total_bill",
                   color="day",
                   pattern_shape="smoker",
                   title="Histogram of Total Bill by Sex, colored by day and patterned by smoker",
                   labels={"total_bill": "Total Bill", "sex": "Sex", "day":"Day", "smoker":"Smoker"},
                   nbins=20 # Adjust number of bins if needed
                   )
fig.show()
```

```python
# prompt: Q4. Using the iris dataset in the Plotly library, Plot a scatter matrix plot, using the "species" column for
# the color parameter.
# Note: Use "sepal_length", "sepal_width", "petal_length", "petal_width" columns only with the
# dimensions parameter.

# Load the iris dataset
iris = px.data.iris()

# Create the scatter matrix plot
fig = px.scatter_matrix(iris,
                        dimensions=["sepal_length", "sepal_width", "petal_length", "petal_width"],
                        color="species")

# Show the plot
fig.show()
```

```python
# Q5. What is Distplot? Using Plotly express, plot a distplot.

# A distplot (short for distribution plot) is a way to visualize the distribution of a dataset  basically, how values are spread out. Traditionally, its a histogram + optional KDE (Kernel Density Estimation) line.
# Plotly Express doesn't have a function called distplot.# However, you can achieve the same thing using px.histogram() with the right options.
import plotly.express as px
import seaborn as sns

# Load tips dataset
tips = sns.load_dataset('tips')

# Create histogram with KDE line (distplot style)
fig = px.histogram(
    tips,
    x='total_bill',
    marginal='rug',         # Adds a rug plot
    nbins=30,
    title='Distribution of Total Bill',
    labels={'total_bill': 'Total Bill'},
    histnorm='density'      # Normalizes the histogram
)

# Show the plot
fig.show()
```
