Q2. Write a python program to print Python Exception Hierarchy.

All exceptions are derived from base class called "Base Exception" and all standard exceptions inherit from this class 

```python
import inspect 

def print_exception_hierarchy(base_exception):
    print(base_exception)
    for subclass in base_exception.__subclasses__():
        print_exception_hierarchy(subclass)

print("exception hierarchy")
print( print_exception_hierarchy(Exception) )
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

lookupError class will be used when the look up operation in the code like getting value for key in a dictionary or extracting particular value from list based on index.
it is the parent class for both KeyError & IndexError. examples below :


```python
# IndexError 

try:
    l= ['a','b','c','d']
    print(l[5])
except IndexError as e:
    print(f"total number of elements in the list is {len(l)} only ")

# keyError
try:
    dict = {'a':1,'b':2,'c':3}
    print(dict['d'])
except KeyError as e:
    print(f"key not found in the dictionary")
```
