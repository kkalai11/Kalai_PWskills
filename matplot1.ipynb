```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
# The subplot() function in Matplotlib is used to create multiple plots in a single figure. It allows you to specify
# the number of rows and columns of subplots, as well as the index of the current subplot. This is useful for comparing

# Use the following data:
import numpy as np
# x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])
# x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])
# x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])
# x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])
import matplotlib.pyplot as plt
import numpy as np
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title("Line Plot 1")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title("Line Plot 2")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title("Line Plot 3")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
plt.title("Line Plot 4")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.tight_layout()
plt.show()
```

```python
# Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
# The bar plot is a type of plot that presents categorical data with rectangular bars. The lengths of the bars are
# proportional to the values they represent. Bar plots are used to compare different categories or groups of data.
# import numpy as np
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

import matplotlib.pyplot as plt
plt.bar(company, profit)
plt.title("Bar Plot")
plt.xlabel("Company")
plt.ylabel("Profit")
plt.show()
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
# The box plot is a standardized way of displaying the distribution of data based on a five-number summary:
# minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum.
# Box plots are used to visualize the spread and skewness of the data, as well as to identify outliers.
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)


import matplotlib.pyplot as plt
import numpy as np

data = [box1, box2]
plt.boxplot(data, labels=["Box 1", "Box 2"])
plt.title("Box Plot")
plt.ylabel("Values")
plt.show()
```
