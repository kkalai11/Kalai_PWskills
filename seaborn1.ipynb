```python
#Que 1: Name any five plots that we can plot using the Seaborn library. Also, state the uses of each plot.
#Ans:
#1. Scatter Plot: Used to visualize the relationship between two continuous variables.
#2. Line Plot: Used to visualize the trend of a variable over time or another continuous variable.
#3. Bar Plot: Used to compare the values of different categories or groups.
#4. Box Plot: Used to visualize the distribution of a continuous variable across different categories.
#5. Heatmap: Used to visualize the correlation between multiple variables in a matrix format.
```

```python
# Que 2: Load the "fmri" dataset using the load_dataset function of seaborn. Plot a line plot using x =
# "timepoint" and y = "signal" for different events and regions.
# Note: timepoint, signal, event, and region are columns in the fmri dataset.
#%pip install seaborn
import seaborn as sns
import matplotlib.pyplot as plt
fmri= sns.load_dataset("fmri")
# Display the first few rows of the dataset
print(fmri.head())
fmri.head()
sns.lineplot(data=fmri,x="timepoint",y="signal",hue="event",style="region")

```

```python
# Que 3: Load the "titanic" dataset using the load_dataset function of seaborn. Plot two box plots using x =
# 'pclass', y = 'age' and y = 'fare'.
# Note: pclass, age, and fare are columns in the titanic dataset.
import seaborn as sns
import matplotlib.pyplot as plt
titanic=sns.load_dataset("titanic")
# Display the first few rows of the dataset
print(titanic.head())   
titanic.head()
sns.boxplot(data=titanic,x="pclass",y="age")
plt.title("Box plot of Age by Passenger Class")
plt.show()
sns.boxplot(data=titanic,x="pclass",y="fare")
plt.title("Box plot of Fare by Passenger Class")
plt.show()
```

```python
# Que 4: Use the "diamonds" dataset from seaborn to plot a histogram for the 'price' column. Use the hue
# parameter for the 'cut' column of the diamonds dataset.
import seaborn as sns
import matplotlib.pyplot as plt
diamonds=sns.load_dataset("diamonds")
# Display the first few rows of the dataset
print(diamonds.head())
diamonds.head()
sns.histplot(data=diamonds,x="price",hue="cut",bins=30,kde=True)
```

```python
# Que 5: Use the "iris" dataset from seaborn to plot a pair plot. Use the hue parameter for the "species" column
# of the iris dataset.
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
# Display the first few rows of the dataset
print(iris.head())
iris.head()
sns.pairplot(data=iris,hue="species")
plt.show()
```

```python
# Que 6: Use the "flights" dataset from seaborn to plot a heatmap.
# Use the pivot_table function to reshape the data for the heatmap.
import seaborn as sns
import matplotlib.pyplot as plt
flights=sns.load_dataset("flights")
# Display the first few rows of the dataset
print(flights.head())
flights.head()
flights_pivot = flights.pivot_table(index='month', columns='year', values='passengers')
plt.figure(figsize=(12, 8))
sns.heatmap(flights_pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap of Passengers by Month and Year")
plt.show()
```
