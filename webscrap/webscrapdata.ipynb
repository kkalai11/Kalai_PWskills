```python
#webscrapping data 
import requests
import bs4
from bs4 import BeautifulSoup as bs
from urllib.request import urlopen
import logging 
```

```python
flipkart_url = 'https://www.flipkart.com/search?q=' +'tv'
```

```python
urlclient=urlopen(flipkart_url)
urlclient
```

```python
# to get all data dump from search page html code"
flipkart_page= urlclient.read()
flipkart_page 
```

```python
# claning using beautifulsoup
flipkart_html= bs(flipkart_page,'html.parser')
```

```python
bigbox=flipkart_html.find_all("div",{"class":"cPHDOP col-12-12"}) # to get all the div with class name _4rR01T_
```

```python
len(bigbox)
del bigbox[0:3] #to remove header data 

product_link='https://www.flipkart.com'+bigbox[0].div.div.div.a['href']
```
